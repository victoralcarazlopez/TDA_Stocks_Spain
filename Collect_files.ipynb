```python
from pandas_datareader import data as pdr
from datetime import date
import yfinance as yf
yf.pdr_override() 
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')
path = 'drive/MyDrive/MUIA/TFM'
```

```python
# Obtener ista de tickers desed fichero

df = pd.read_excel(f'{path}/data/tickers.xlsx') 
ticker_list = df.iloc[:,0].tolist()
print(ticker_list)
```

```python
start_date= "2000-01-01"
end_date="2023-03-21"

files = {}


# Recolectar datos de cada empresa
for ticker in ticker_list:
  print(ticker)
  data = pdr.get_data_yahoo(ticker, start=start_date, end=end_date)
  files[ticker] = data
```

```python
files['ACS.MC']
```

```python
for ticker in files:
  df = files[ticker]
  df.to_csv(f'{path}/data/Cotizaciones_RAW/{ticker}.csv')
  
```
